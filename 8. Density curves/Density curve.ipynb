```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

sns.set(style = "whitegrid")
```

```python
# Creating a sample dataset 
np.random.seed(42)

scores = np.random.normal(loc= 70, scale= 10, size= 200)
df = pd.DataFrame({"Scores" : scores})
df.head()
```

```python
plt.figure()
plt.hist(df['Scores'], bins= 20)
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.title("Histogram of scores")
```

```python
# KDE PLOT 
plt.figure() 
sns.kdeplot(df["Scores"], fill=True)
plt.xlabel("Score")
plt.ylabel("Density")
plt.title("Density curve (KDE) of scores")
```

```python
plt.figure()
sns.histplot(df["Scores"], bins=20, kde=True)
plt.xlabel("Score")
plt.ylabel("Count")
plt.title("Histogram with Density Curve")
```

```python
mean = df["Scores"].mean()
median = df["Scores"].median()
mode = df["Scores"].mode()[0]

plt.figure()
sns.kdeplot(df["Scores"], fill=True)
plt.axvline(mean, linestyle="--", label=f"Mean: {mean:.2f}")
plt.axvline(median, linestyle="-", label=f"Median: {median:.2f}")
plt.axvline(mode, linestyle="-.", color="blue", label=f"Mode: {mode:.2f}") # Mode line
plt.legend()
plt.title("Density Curve with Mean & Median")
```

```python
skewed_data = np.random.exponential(scale=50, size=300)
df2 = pd.DataFrame({"Income": skewed_data})

plt.figure()
sns.histplot(df2["Income"], kde=True)
plt.xlabel("Income")
plt.title("Right-Skewed Distribution (Income-like Data)")
plt.show()
```

```python
plt.figure()
sns.kdeplot(df["Scores"], label="Exam Scores", fill=True)
sns.kdeplot(df2["Income"], label="Income", fill=True)
plt.legend()
plt.title("Comparison of Density Curves")
```
